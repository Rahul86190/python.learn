## scatter plot

it shows thebullet points  . a single point represent as (x,y)

```python
ls1=[25,35,68,24,61]
ls2=[96,45,12,36,45]
plt.scatter(ls1,ls2)
plt.xlabel("Ls1")
plt.ylabel("Ls2")
plt.title("Scatter plot")
plt.show
```

```python
plt.scatter(ls1,ls2,marker="*",s=600)  # s for size of marker
```

```python
plt.scatter(ls1,ls2,marker="v",color="red",s=600)
plt.show
```

## HISOGRAM

It gives information about the distribution of data

```python
age=np.random.randint(1,100,1000)
```

```python
plt.hist(age,ec="k")
plt.show()
```

```python
boxex_dist=[0,18,30,50,70,90,110]       # By using this bins we can  create a range like here 0-18,18-30,30-50,50-70,70-90,90-110
plt.hist(age,bins=boxex_dist,ec="k")
plt.show()
```
