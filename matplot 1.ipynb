# Linechart

When data of both axis is numerical there use Linechart

```python
ls=[41,52,65,60]
plt.plot(ls)
```

```python
ls1=[41,52,65,60]
ls2=[25,63,45,75]
plt.plot(ls,ls2)
```

```python
ls1=[41,52,65,60]
ls2=[25,63,45,75]
plt.plot(ls,ls2)
plt.xlabel("LS-1")
plt.ylabel("LS-2")
```

```python
ls1=[41,52,65,60]
ls2=[25,63,45,75]
plt.plot(ls1,ls2,marker ="*",ms=10)
```

```python
arr=np.random.randint(1,200,200)
plt.plot(arr)
```

# Bar plot

If one data is categorical(alphabet) and second is numerical then use Bar plot

```python
country_name=["Japan","India","Itli","Bangladesh"]
country_population=[400,900,466,370]

plt.xlabel("Country")
plt.ylabel("popolation in crore")
plt.bar(country_name,country_population)
```

Working on real data 

```python
df=pd.read_csv("Used_Bikes.csv")
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["owner"].value_counts()
```

```python
df["owner"].value_counts().keys()
```

```python
ownership=(df["owner"].value_counts().keys())
ownership
```

```python
no_of_bikes=list(df["owner"].value_counts().values)
no_of_bikes
```

```python
plt.bar(ownership,no_of_bikes)
plt.xlabel("ownership")
plt.ylabel("no of bikes")
plt.title("Ownership chart")
```

```python
df["brand"].value_counts()
```

```python
brand_names=list(df["brand"].value_counts().keys())
brand_names
```

```python
brand_count=list(df["brand"].value_counts().values)
```

```python
plt.bar(brand_names,brand_count,color="red",ec="k")   # ec = edge color  and b=blue,k = black  , lw = line width
plt.xlabel("Brnads")
plt.ylabel("store of bikes")
plt.xticks(rotation=90)
plt.yticks(rotation=60)
plt.title("Brands available")
```

```python
plt.bar(brand_names,brand_count,color="green",ec="k",lw=2,ls="dotted")   # ec = edge color  and b=blue,k = black  , lw = line width, ls= line style
plt.xlabel("Brnads")
plt.ylabel("store of bikes")
plt.xticks(rotation=90)
plt.yticks(rotation=60)
plt.title("Brands available")
```

# pie plot

```python
print(country_name)
print(country_population)
plt.pie(country_population,labels=country_name)
plt.show()
```

```python
plt.pie(country_population,labels=country_name,autopct="%2.4f%%")  #autopct= autopercentage
plt.show()
```

```python
plt.pie(country_population,labels=country_name,autopct="%2.4f%%",wedgeprops={'ec':"black","ls":"dashed","lw":1})  #autopct= autopercentage
plt.figure(figsize=(8,8))
plt.show()
```

```python
explosion=[0,0,0,0]
plt.pie(country_population,labels=country_name,autopct="%2.2f%%",wedgeprops={'ec':"black",'ls':'dashed','linewidth':1},explode=explosion)
plt.show()
```

```python
explosion=[0,0,0.2,0]
plt.pie(country_population,labels=country_name,autopct="%2.2f%%",explode=explosion)
plt.show()
```

```python
explosion=[0,0,0.2,0.1]
plt.pie(country_population,labels=country_name,autopct="%2.2f%%",explode=explosion)
plt.show()
```

```python
explosion=[0,.2,0,0]
plt.pie(country_population,labels=country_name,autopct="%2.2f%%",wedgeprops={"ec":"black"},explode=explosion)
plt.show()
```

```python
print(country_name)
print(country_population)
explosion=[0,0,0.2,0]
print(explosion)
plt.figure(figsize=(8,8))
plt.pie(country_population,labels=country_name,autopct="%2.3f%%",wedgeprops={"ec":"black"},explode=explosion,shadow="green")
plt.savefig("country_data")
plt.show()
```
